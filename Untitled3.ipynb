```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('housing.csv')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
x
```

```python
y
```

```python
#create are test data i.e for like 20% of the actual data 
#it for later use after the training and hyper parameters tunning of the data before launching to evaluate the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
train_data
```

```python
sns.pairplot(train_data)
```

```python
#get multiple histograms showing the distribution of various features and their correlation to the target variable
train_data.hist(figsize=(15,10))
```

```python
#to visualize a  correlation matrix we use a heatmap
#A correlation matrix is simply a table which displays the correlation coefficients for different variables. The matrix depicts the correlation between all the possible pairs of values in a table.
train_data.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
#PREPROCESSING STAGE
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,10))
```

```python
#assumption is that ocean proximity will influence the price 
#so the values placed for ocean proximity should be changed to numerical values. a more binary value
train_data.ocean_proximity.value_counts()# just for checking 
pd.get_dummies(train_data.ocean_proximity)#just for checking
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
#this correlation matrix will show the correlation of pricing and the distance from the coast
plt.figure (figsize=(15,10))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
#use scatter plot to show the change of prices as they move further of the coast
#houses nearer the ocean are of the hue palrttr of red showing that their more expensive.
plt.figure (figsize=(15,10))
sns.scatterplot(x='latitude',y='longitude',data=train_data,hue = 'median_house_value',palette='coolwarm')
```

```python
#Feature engineering
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_bedrooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
plt.figure(figsize=(15,10))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
#training multiple models using linear regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
x_train_s=scaler.fit_transform(x_train)

reg=LinearRegression()
reg.fit(x_train_s, y_train)
```

```python
test_data=x_test.join(y_test)
test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
test_data['bedroom_ratio']=test_data['total_bedrooms']/test_data['total_bedrooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']
x_test,y_test=test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```
